```python
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dep=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(indep,dep,test_size=0.30, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_trainSC=sc.fit_transform(X_train)
#X_test=sc.transform(X_test)
```

```python
X_trainSC
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
grid_params={'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'gamma':['scale', 'auto'],'C':[100,200,300,400,500]}
grid=GridSearchCV(SVR(), grid_params, verbose=3, n_jobs=-1,refit=True)
grid.fit(X_trainSC, Y_train)
```

```python
from sklearn.metrics import r2_score
grid_prediction=grid.predict(X_test)
r_score=r2_score(Y_test, grid_prediction)
re=grid.cv_results_
print('The R_Score of the best parameter{}:'.format(grid.best_params_))
```
